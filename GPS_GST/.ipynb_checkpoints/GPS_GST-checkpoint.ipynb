```python
import pandas as pd
#pip install dataframe_image
import dataframe_image as dfi
```

```python
basePrice = 3000
def GetTotalAmt(amt,gst=0.18):
    return round(amt+(amt*gst))
    
def GetTotalAmount(unitPrice,qty=1):
    gst_df = pd.DataFrame()
    gst_df["Quantity"] = [x for x in range(1,qty+1)]
    gst_df["Unit Price"] = [unitPrice for x in range(1,qty+1)]
    gst_df["Pre Tax Total"] = [unitPrice*x for x in range(1,qty+1)]
    gst_df["GST"] = ["18%" for x in range(1,qty+1)]
    gst_df["Total"] = [GetTotalAmt(unitPrice*x) for x in range(1,qty+1)]
    gst_df["Profit"] = [(unitPrice-basePrice)*x for x in range(1,qty+1)]
    return gst_df
```

```python
my_df = GetTotalAmount(3400,23)
blankIndex=[''] * len(my_df)
my_df.index=blankIndex
my_df.to_csv('MyProfitAnalysis.csv')
dfi.export(my_df,'MyProfit.png')
my_df.tail()
```
